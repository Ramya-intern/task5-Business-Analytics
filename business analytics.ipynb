```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
dataset=pd.read_csv("C:\\Users\\Ramya\\Desktop\\Untitled Folder\\SampleSuperstore.csv")
```

```python
dataset.info()
```

```python
dataset.shape
```

```python
# checking whether there are any null values
dataset.isnull().sum()
```

```python
#Checking if any duplicate row is there in the dataset
dataset.duplicated().sum()
```

```python
#Removing 17 duplicated rows to avoid faults in further calculation
dataset.drop_duplicates(inplace=True)
```

```python
#removing the unimportant columns
dataset=dataset.drop(['Country', 'Postal Code'], axis=1)
```

### Data Visualization

```python
#Category wise sales in Each region
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Region", y="Sales", hue="Category", data=dataset, palette="PuBu")
```

#### The South region is leading the sales followed by East,Central and west.

#### Technology sales are higher in all the categories followed by furniture.

```python
#Segment vise sales in Each region
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Region", y="Sales", hue="Segment", data=dataset, palette="OrRd")
```

#### No much difference in sales for each region.

```python
# Profit calculation for the categories
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Category", y="Profit", data=dataset, palette="binary")
```

#### from the above visualisation it can be see that profit is high in Technology Category.

```python
#Category wise Profit in Each Region
plt.figure(figsize=[12,8])
ax = sns.barplot(x="Region", y="Profit", hue="Category", data=dataset, palette="YlOrRd")
```

#### Observation:- Simply 'Technology' generates highest profit in every region. And 'Furniture' is lagging in all the regions except South region. Most importantly, at Central region, 'Furniture' is experiencing loss. These points are to be noted and taken care of.

```python
#Subcategory wise profit
plt.figure(figsize=[10,8])
ax = sns.barplot(x="Sub-Category", y="Profit", data=dataset, palette="magma")
plt.xlabel("Subcategory", fontsize=15)
plt.ylabel("Profit",fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
#Entries with Category=Technology
cat=dataset[(dataset['Category']=="Technology")]
```

#### 'Copiers' are gaining huge profit. 'Accessories' are also doing good. But 'Tables' and 'Bookcases' are going with loss. Respective steps are to be taken to improve the business in these Subcategories.

```python
#Let's get the Sales of each Subcategory under Technology
plt.figure(figsize=[20,15])
ax = sns.barplot(x="Sub-Category", y="Sales", data=cat, palette="PiYG")
plt.xlabel("Subcategories",fontsize=5)
plt.ylabel("Sales",fontsize=5)
```

#### In category 'Technology', subcategory 'Copiers' is having the highest sales, where 'Accessories' is having lowest amount of sales. 'Machines' subcategory is also performing good.¶
